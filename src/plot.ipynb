```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("inputs/dataset1-7000.csv")
# List all column names in the dataset
# print(data.columns)

attributes = ['gender', 'exhaust_volume', 'motorcycle_style', 'passenger_on_backseat',
       'helmet', 'height', 'weight', 'age', 'purchase_price', 'monthly_mileage', 'traffic_ticket_count',
       'speeding_habbit', 'has_license', 'has_modifications',
       'is_noisy_engine', 'is_mountain_ride', 'has_regular_maintenance',
       'commute_with_motorcycle']

```

```python
for attr in attributes:
    plt.figure(figsize=(6, 3))
    sns.histplot(data[attr], bins=20)
    plt.title(attr + " Distribution")
    plt.tick_params(axis='x', labelsize = 6)
    plt.xlabel(attr)
    plt.show()
```
